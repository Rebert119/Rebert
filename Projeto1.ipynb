```python
import pandas as pd

df = pd.read_parquet('db.parquet')

df['DataCarga'] = pd.to_datetime(df['DataCarga'])
df = df.sort_values(by=['Tabela', 'DataCarga'])

df['Crescimento_Absoluto'] = df.groupby('Tabela')['EspacoUsadoKB'].diff()

stats = df.dropna(subset=['Crescimento_Absoluto']).groupby('Tabela')['Crescimento_Absoluto'].agg(['mean', 'std']).reset_index()

df = df.merge(stats, on='Tabela', how='left')

df['std'] = df['std'].fillna(0)
df['Anomalia'] = df['Crescimento_Absoluto'] > (df['mean'] + 3 * df['std'])

alertas = df[df['Anomalia'] == True]

display(df.head(10000))
```

```python
tabelas_com_mudanca = df.groupby('Tabela')['EspacoUsadoKB'].agg(['min', 'max']).reset_index()
tabelas_com_mudanca = tabelas_com_mudanca[tabelas_com_mudanca['min'] != tabelas_com_mudanca['max']]

print("Tabelas que tiveram crescimento real no arquivo:")
display(tabelas_com_mudanca)

nomes_tabelas = tabelas_com_mudanca['Tabela'].tolist()
df_crescimento = df[df['Tabela'].isin(nomes_tabelas)].copy()

df_crescimento['Taxa_Crescimento_Pct'] = df_crescimento.groupby('Tabela')['EspacoUsadoKB'].pct_change() * 100
display(df_crescimento[['DataCarga', 'Tabela', 'EspacoUsadoKB', 'Taxa_Crescimento_Pct']].dropna())
```

```python
import plotly.express as px
top_5_crescimento = df.groupby('Tabela')['Crescimento_Absoluto'].sum().nlargest(5).index

df_top5 = df[df['Tabela'].isin(top_5_crescimento)]
df_top5_plot = df_top5.groupby(['DataCarga', 'Tabela'])['EspacoUsadoKB'].max().reset_index()

fig = px.line(df_top5_plot, x='DataCarga', y='EspacoUsadoKB', color='Tabela', title='Top 5 Tabelas em Crescimento')
fig.show()
```

```python
import pandas as pd
import numpy as np
import plotly.express as px

pd.options.display.float_format = '{:.2f}'.format

df = pd.read_parquet('db.parquet')
df['DataCarga'] = pd.to_datetime(df['DataCarga'])

df = df.groupby(['Tabela', 'DataCarga'])['EspacoUsadoKB'].max().reset_index()
df = df.sort_values(by=['Tabela', 'DataCarga'])

df['Crescimento_Absoluto'] = df.groupby('Tabela')['EspacoUsadoKB'].diff()
df['Taxa_Crescimento_Pct'] = df.groupby('Tabela')['EspacoUsadoKB'].pct_change() * 100
```

```python
ultima_data = df['DataCarga'].max()
p1 = df[df['DataCarga'] == ultima_data].nlargest(5, 'EspacoUsadoKB')
print("P1: Tabelas com maior volume atual (KB):")
display(p1[['Tabela', 'EspacoUsadoKB']])
```

```python
p2 = df.groupby('Tabela')['Crescimento_Absoluto'].sum().nlargest(5).reset_index()
print("\nP2: Tabelas que mais ganharam tamanho (Total acumulado em KB):")
display(p2)
```

```python
# Média histórica de crescimento
media_hist = df.groupby('Tabela')['Crescimento_Absoluto'].mean().reset_index(name='Media_Historica')
# Crescimentoo anual
atual = df.sort_values('DataCarga').groupby('Tabela').tail(1)[['Tabela', 'Crescimento_Absoluto']]
# Novas Colunas
p4 = atual.merge(media_hist, on='Tabela')
p4['DiferencaAtual_vs_Media'] = p4['Crescimento_Absoluto'] - p4['Media_Historica']
print("\nP4: Crescimento Atual vs Média Histórica (Comparação de períodos):")
display(p4.nlargest(10, 'Crescimento_Absoluto'))
```

```python
stats = df.groupby('Tabela')['Crescimento_Absoluto'].agg(['mean', 'std']).reset_index()
df_stats = df.merge(stats, on='Tabela')
df_stats['Anomalia'] = df_stats['Crescimento_Absoluto'] > (df_stats['mean'] + 2 * df_stats['std'])

print("\nP5: Alertas de crescimentos fora do padrão histórico:")
display(df_stats[df_stats['Anomalia'] == True][['DataCarga', 'Tabela', 'Crescimento_Absoluto', 'mean', 'std']].head())
```

```python
top_tabelas = p1['Tabela'].head(5).tolist()
df_plot = df[df['Tabela'].isin(top_tabelas)]

fig = px.line(df_plot, x='DataCarga', y='EspacoUsadoKB', color='Tabela',
              title='Tendência de Crescimento das Principais Tabelas',
              template='plotly_white')
fig.update_layout(yaxis_title="Espaço em KB", xaxis_title="Data da Coleta")
fig.show()
```
